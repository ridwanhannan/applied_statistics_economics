# Applied Statistics for Economics

T is the cigarette tax (T=0 means “No tax” and T=4 means “$4/pack-tax”), H is a health-risk index (H=0 means “Low risk” and H=5 means “High risk”), and the figures in the table are the known joint probabilities:


[11] 1. Determine the standard deviation of the health risk index, sd(H). [Note: This is not a conditional standard deviation.]

[22] 2. Determine the variance of the health risk index given that the cigarette tax is $4/pack, i.e. determine var(H|T=4).  [Hint: This is a conditional variance.]

[11] 3. Determine the correlation between the health risk index and the cigarette tax, i.e. corr(T, H). 


```r
# Function to calculate statistics for the joint probability distribution 

statistics_summary <- function(H0_T0, H5_T0, H0_T4, H5_T4) {
  # marginal probabilities 
  marginal_H0 = H0_T0 + H0_T4 
  marginal_H5 = H5_T0 + H5_T4 
  marginal_T0 = H0_T0 + H5_T0
  marginal_T4 = H0_T4 + H5_T4
  
  # Expected Values 
  Expected_H = (0 * marginal_H0) + (5 * marginal_H5)
  Expected_T = (0 * marginal_T0) + (4 * marginal_T4)
  
  # Variances
  Var_H = ((0 - Expected_H)**2 * marginal_H0) + ((5 - Expected_H)**2 * marginal_H5)
  Var_T = ((0 - Expected_T)**2 * marginal_T0) + ((4 - Expected_T)**2 * marginal_T4)
  
  
  # Standard Deviation of H 
  stdev = sqrt(Var_H)
  print(paste("The Standard Deviation of the Health Risk Index is ", stdev, "units"))

  # Variance of H given T = 4
  pr_h_0_given_t_4 = H0_T4 / marginal_T4
  pr_h_5_given_t_4 = H5_T4 / marginal_T4
  Expected_H_given_T_4 = ((pr_h_0_given_t_4) * 0) + ((pr_h_5_given_t_4) * 5) 
  Var_H_given_T_4 = (((0 - Expected_H_given_T_4)**2) * pr_h_0_given_t_4) + (((5 - Expected_H_given_T_4)**2) * pr_h_5_given_t_4)
  print(paste("The variance of the health risk index given that the cigarette tax is $4/pack is ", Var_H_given_T_4, "units"))
  
  # Correlation between Health Risk Index and Cigarrette Tax 
  cov_T_H = H0_T0 * (0 - Expected_T) * (0 - Expected_H) + 
    H5_T0 * (0 - Expected_T) * (0 - Expected_H) +
    H0_T4 * (4 - Expected_T) * (5 - Expected_H) +
    H5_T4 * (4 - Expected_T) * (5 - Expected_H)
  
  corr_T_H = cov_T_H / (sqrt(Var_T * Var_H))
  print(paste("The correlation between the health risk index and the cigarette tax is ", corr_T_H))
}

statistics_summary(0.20, 0.40, 0.25, 0.15)
```

A strong positive correlation between health index and cigarrete tax suggests that the two variables move together in the same direction. That is the higher the tax on cigarrettes the higher the health risk index which means there is a higher health risk. This makes sense as the government would want to decentivise purchases of higher health risk index cigarettes by increasing the taxes on those products.
